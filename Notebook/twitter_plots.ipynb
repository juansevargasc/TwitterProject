## Analyze

```python
exclude_words = ['ha', 'est치n', 'nuestra', 'm치s', 'fue', 'este', '#lom치strinado', '#entretenimiento', '#ojodelanoche', '#atenci칩n', 'ser', '#deportes', 'medio', 'entre', '#ampliaci칩n', 'desde', 'durante','#video', 'as칤', '游녢','#lo칰ltimo', 'minuto', 'hay', 'tras', '&gt;&gt;&gt;', '&gt;&gt;', 'qu칠', 'vea','쯤u칠', 'vivo', 'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'al', 'del', 'lo', 'le', 'y', 'e', 'o', 'u', 'de', 'a', 'en', 'que', 'es', 'por', 'para', 'con', 'se', 'su', 'les', 'me', 'q', 'te', 'pero', 'mi', 'ya', 'cuando', 'como', 'estoy', 'voy', 'porque', 'he', 'son', 'solo', 'tengo', 'muy', '-', 'l', '|' , 'sobre', 'sus', 'est치', 'esta', ' vivo']


def analyze(topic):
    top_words = {}
    tweets_topic = open(topic, encoding='utf-8')
    for line in tweets_topic:
        words = line.strip().lower().split()
        for word in words:
            if word not in exclude_words:
                top_words[word] = top_words.get(word, 0) + 1 # get method returns value for word. If word does not exists, returns 0. If value is not specified, returns None, that's why I've set 0.

    
    tweets_topic.close()
    return top_words


def most_mentioned_people(top_words, most_used_words):
    mentioned = {}
    count_u = 0
    for word in most_used_words:
        if count_u < 10 and word.startswith('@'):
            print(top_words[word], word)
            mentioned[word] = top_words[word]
            count_u += 1
    print('*' * 40) # esto es para separar visualmente la informaci칩n
    return mentioned

def top20_words(top_words, most_used_words):
    the20 = {}
    count = 0
    for word in most_used_words:
        if count < 35:
            print(top_words[word], word)
            the20[word] = top_words[word]
            count += 1
    return the20
     
 
```

**Ejemplo con #ParoIndefinido2021**

```python
top = analyze('#ParoIndefinido2021.txt') # Returns array of top words
top_sorted = sorted(top, key = top.get, reverse = True) # Orders by quantity of words

#So Results:
most_mentioned_people(top, top_sorted)
top20_words(top, top_sorted)
```

## Analyzing Sentences

```python
exclude_words = ['ha', 'est치n', 'nuestra', 'm치s', 'fue', 'este', '#lom치strinado', '#entretenimiento', '#ojodelanoche', '#atenci칩n', 'ser', '#deportes', 'medio', 'entre', '#ampliaci칩n', 'desde', 'durante','#video', 'as칤', '游녢','#lo칰ltimo', 'minuto', 'hay', 'tras', '&gt;&gt;&gt;', '&gt;&gt;', 'qu칠', 'vea','쯤u칠', 'vivo', 'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'al', 'del', 'lo', 'le', 'y', 'e', 'o', 'u', 'de', 'a', 'en', 'que', 'es', 'por', 'para', 'con', 'se', 'su', 'les', 'me', 'q', 'te', 'pero', 'mi', 'ya', 'cuando', 'como', 'estoy', 'voy', 'porque', 'he', 'son', 'solo', 'tengo', 'muy', '-', 'l', '|' , 'sobre', 'sus', 'est치', 'esta', ' vivo']


def analyze(topic):
    count_v = 0
    count_CAI = 0
    count_abuso = 0
    count_derechos = 0

    top_words = {}
    tweets_topic = open(topic, encoding='utf-8')

    for line in tweets_topic:
        if 'v치ndalo' in line.lower() or 'v치ndalos' in line.lower():
          count_v += 1
        if 'cai' in line.lower() or '(cai)' in line.lower():
          count_CAI += 1
        if 'abuso policial' in line.lower() or 'abusos' in line.lower():
          count_abuso += 1
        if 'derechos humanos' in line.lower() or 'derecho humano' in line.lower():
          count_derechos += 1

    


    tweets_topic.close()
    return count_v, count_CAI, count_abuso, count_derechos


def most_mentioned_people(top_words, most_used_words):
    mentioned = {}
    count_u = 0
    for word in most_used_words:
        if count_u < 10 and word.startswith('@'):
            print(top_words[word], word)
            mentioned[word] = top_words[word]
            count_u += 1
    print('*' * 40) # esto es para separar visualmente la informaci칩n
    return mentioned

def top20_words(top_words, most_used_words):
    the20 = {}
    count = 0
    for word in most_used_words:
        if count < 35:
            print(top_words[word], word)
            the20[word] = top_words[word]
            count += 1
    return the20

def count_tweets(topic):
    count = 0
    tweets_topic = open(topic, encoding='utf-8')

    for line in tweets_topic:
        if ( not line.startswith('\n') ) :
          count += 1
          if count <= 4:
            print(line)

    tweets_topic.close()
    return count

def count_tweets_per_day(topic, date):
    count = 0
    tweets_topic = open(topic, encoding='utf-8')

    for line in tweets_topic:
        if date in line:
          count += 1

    tweets_topic.close()
    return count
  

def count_word(topic, the_word):
  count = 0
  tweets_topic = open(topic, encoding='utf-8')

  for line in tweets_topic:
      if the_word.lower() in line.lower():
        count += 1
        print(line)

  tweets_topic.close()
  return count
```

```python
cols = ['c','y','r','g']
counts = analyze('NoticiasRCN.txt')
categorias = ['V치ndalo', 'CAI', 'Abuso policial', 'Derechos Humanos']

counts
```
